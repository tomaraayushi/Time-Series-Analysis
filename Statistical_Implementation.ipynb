**Measure of Central Tendency**

2.Median

3.Mode

```python
ages = [23, 24, 32, 45, 12, 43, 67, 45, 32, 56, 32, 120]
```

```python
import numpy as np
```

```python
print(np.mean(ages))
print(np.median(ages))
```

```python
import statistics
print(statistics.mean(ages))
print(statistics.median(ages))
```

```python
import seaborn as sns
sns.boxplot(ages)
```

```python
# Finding outliers
q1, q3 = np.percentile(ages, [25, 75])
```

```python
# to check outlier [lower fence - higher fence]
IQR = q3 - q1
lower_fence = q1 - 1.5*IQR
higher_fence = q3 + 1.5*IQR
print(lower_fence, higher_fence)
```

#Measure of Dispersion#

1. Variance

2. Standard Deviation


```python
# statistics calculate variance for sample data
statistics.variance(ages)
```

```python
# np calculate variance for whole population
np.var(ages,axis=0)
```

```python
## Population variance
def p_variance(data):
  n = len(ages)
  ## mean of data
  mean = sum(data)/n
  ## variance
  deviation = [(x-mean)**2 for x in data]
  p_var = sum(deviation)/n
  return p_var

p_variance(ages)
```

```python
## sample variance
def s_variance(data):
  n = len(ages)
  mean = sum(data)/n
  deviation = [(x-mean)**2 for x in data]
  s_var = sum(deviation)/(n-1)
  return s_var

s_variance(ages)
```

```python
## pvariance
statistics.pvariance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

#Check whether a distribution is NORMAL

```python
## Q-Q Plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
  plt.figure(figsize = (10,6))
  plt.subplot(1,2,1)
  sns.histplot(sample)
  stat.probplot(sample, dist= 'norm', plot = pylab)
  plt.show()
```

```python
# create a normal distributed data
s = np.random.normal(0.5, 0.2, 1000)
plot_data(s)
```
